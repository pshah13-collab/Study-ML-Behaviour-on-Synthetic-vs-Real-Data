```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
path = "/Users/skush/CodeX/Stress-Level-Classifier/data/processed/survey_data_labelled.csv"
```

```python
df = pd.read_csv(path)
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
```

```python
df.columns 
```

<h3> Irrelevant Columns

```python
irrelevant = []
```

```python
df = df.drop(columns=irrelevant)
```

<h3> Encoding Target Variable </h3>

```python
le = LabelEncoder()
df['Stress_Level'] = le.fit_transform(df['Stress_Level'])
```

```python
df.Stress_Level.value_counts()
```

```python
X = df.drop('Stress_Level', axis=1)
y = df['Stress_Level']
```

```python
X = pd.get_dummies(X, drop_first=True)
```

<h3> Train Test Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.3,
    stratify=y,
    random_state=42
)
```

<h3> Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(
    sampling_strategy={0: 100, 1: 100, 2: 100},  # <-- specify per class
    k_neighbors=2,
    random_state=42
)
```

```python
X_train_smote, y_train_smote = smote.fit_resample(X_train_scaled, y_train)
```

```python
X_train_scaled.shape
```

```python
X_train_smote.shape
```

<h3> Random Forest Training 

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, accuracy_score

knn_model = KNeighborsClassifier(n_neighbors=5)

knn_model.fit(X_train_smote, y_train_smote)
```

```python
y_pred_knn = knn_model.predict(X_test_scaled)
```

<h3> Model Evaluation 

```python
print("KNN Accuracy:", accuracy_score(y_test, y_pred_knn))
print(classification_report(y_test, y_pred_knn))
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, y_pred_knn)

plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")

plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix - Logistic Regression")

plt.show()
```
